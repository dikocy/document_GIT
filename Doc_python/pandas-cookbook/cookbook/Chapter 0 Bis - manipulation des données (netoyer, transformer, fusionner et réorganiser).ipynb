```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

* pandas.merge : connecte les lignes de plusieurs DataFrame sur la base d'une ou plusieurs clé(s).
* pandas.concat : colle ou empile des objets ensemble le long d'un axe.
* La méthode 'conbine_first' coupe (slicing) ensemble des données superposées pour remplir les vecteurs
  manquantes dans un objet avec les valeurs d'un autre.

```python
df1 = DataFrame({'key':['b','b','a','c','a','a','b'], 'data1':range(7)})
df2 = DataFrame({'key': ['a','b','d'], 'data2':range(3)})
df1
```

```python
df2
```

```python
pd.merge(df1,df2, on='key')
```

Si les noms de colonnes sont différents dans chaque objet, on peut les spécifier séparement:

```python
df3 = DataFrame({'lkey':['b','b','a','c','a','a','b'], 'data1':range(7)})
df4 = DataFrame({'rkey': ['a','b','d'], 'data2':range(3)})
pd.merge(df3,df4, left_on='lkey', right_on='rkey')
```

Par defaut 'merge' réalise une jointure 'inner' (interne); les clés du résultat sont l'intersection.
il existe d'autre fonctions:
* 'left' (gauche)
* 'right' (droit)
* 'outer' (externe)
La jointure 

```python
pd.merge(df1, df2, how = 'right')
```
